```python
import pandas as pd
```

```python
pairs = pd.read_csv("transliteration/Hindi - Word Transliteration Pairs 1.txt",sep="\t", header=None)
```

```python
pairs.columns = ["roman", "hindi"]
```

```python
pairs.dropna(inplace=True)
```

```python
pairs.reset_index(inplace=True)
```

```python
pairs.head(10)
```

```python
temp = dict(pairs.groupby('hindi')['roman'].apply(list))
```

```python
inv_map = dict()
for value in temp.values():
    inv_map[value[0]] = value[0:]
```

```python
inv_map
```

```python
def invert_dict(d): 
    inverse = dict() 
    for key in d: 
        # Go through the list that is saved in the dict:
        for item in d[key]:
            # Check if in the inverted dict the key exists
            if item not in inverse: 
                # If not create a new list
                inverse[item] = key
#             else: 
#                 inverse[item].append(key) 
    return inverse
```

```python
mapper = invert_dict(inv_map)
```

```python
mapper
```

```python
import numpy as np

# Save
np.save('normalization/mapper.npy', mapper) 

# # Load
# read_dictionary = np.load('my_file.npy').item()
# print(read_dictionary['hello']) # displays "world"
```
